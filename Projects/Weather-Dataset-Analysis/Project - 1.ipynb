# Q.11. What is the Mean value of each column against each 'Weather Condition'?

```python
data.head(1)
```

```python
data.groupby("Weather Condition").mean(numeric_only=True)
```

```python
'''If we don't write (numeric only = True) in the bracket of mean, then it will show errors, because the Date/Time and Weather Condition columns are strings. That's why the mean() function will not
be able to calculate the mean of the whole dataset. 

To avoid this error, we must put the mean() function as mean(numeric_only = True).'''
```

# Q.14. Find all instances when weather is "Clear" or visibility is above 40.

# Q.15. Find all instances when:

A. 'Weather Condition' is 'Clear' and 'Rel Hum_%' is greater than 50.

or, 

B. 'Visibility_km' is above 40.

```python
data[(data["Weather Condition"] == "Clear") & (data["Rel Hum_%"] > 50)]
```

```python
# Answer: 1070 times
```

```python
data[data["Visibility_km"] > 40]
```

```python
# Answer: 2014 times
```
